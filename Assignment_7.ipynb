#### 1. Create an sqlalchemy engine using a sample from the data set

```python
# import libraries
import sqlalchemy
import numpy as np
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy import Sequence 
Column('id', Integer, Sequence('user_id_seq'), primary_key=True)
from sqlalchemy.sql import select,insert,delete

# create sqlalchemy engine
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:', echo=True)


from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
Base.metadata.create_all(engine)

# create sample dataset table
class Mydb(Base):
            __tablename__ = 'mydb' 
        
            id = Column('id',Integer,Sequence('user_id_seq'), primary_key=True)
            age = Column(Integer)
            workclass = Column(String)
            education = Column(String)
            marital_status = Column(String)
            occupation = Column(String)
            native_country = Column(String)
            def __repr__(self):
                 return "<User(age = '%s', workclass = '%s', education = '%s', marital_status = '%s',occupation = '%s', native_country = '%s')>" 
                 % (self.age, self.workclass, self.education,self.marital_status,self.occupation,self.native_country)

Mydb.__table__
```

```python
Base.metadata.create_all(engine)

# import libraries
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf

# pandas sql dataframe
pysqldf = lambda q: sqldf(q, globals())

# read dataset
sqladb = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',names=['Age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','>50K, <=50K'])

# connect to the sqlalchemy engine
conn = engine.connect()

from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
Session = sessionmaker()
Session.configure(bind=engine)
```

```python
session = Session()

for i in np.array(sqladb[['Age','workclass','education','marital-status','occupation','native-country']].tail(10)) :
        rec_add1 = Mydb(age = i[0], workclass = i[1], education = i[2], marital_status = i[3], occupation = i[4], native_country = i[5])
        session.add(rec_add1)

session.commit()
```

```python
from sqlalchemy.sql import select
s = select([Mydb])
result = conn.execute(s)

# sample dataset
for row in result:
    print(row)
```
